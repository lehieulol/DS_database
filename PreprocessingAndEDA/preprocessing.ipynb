```python
import pandas as pd
import glob
import os 
import json
```

gộp data ở 3 file thành 1 file csv

```python
current_directory = os.path.dirname(os.path.realpath('preprocessing.ipynb'))
database_directory=os.path.join(current_directory,'..',"Database")
os.chdir(database_directory)

extension='txt'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

data_list=[]

for file in all_filenames:
    with open(file,'r') as f:
        file_content=f.read()
        for line in file_content.split('\n'):
            if line.strip(): 
                line=line.replace("'","\"")
                data_dict=json.loads(line)
                data_list.append(data_dict)

combined_data=pd.DataFrame(data_list)

combined_data.to_csv("combined_data.csv", index=False, encoding='utf-8-sig')

```

```python
important_columns = ['brand','dtime', 'flytime','atime','price', 'dcity', 'acity', 'flight_type', 'class_type']
df_no_duplicates = df.drop_duplicates(subset=important_columns, keep='first')
num_removed_rows = len(df) - len(df_no_duplicates)
print(f"Removed {num_removed_rows} duplicate rows.")
print(df_no_duplicates)
```

```python
#check for duplicated rows
duplicated_rows = df.duplicated()

if duplicated_rows.any():
    duplicated_data = df[duplicated_rows]
    print("Duplicated Rows:")
    print(duplicated_data)

    #drop duplicated rows and keep the first occurrence
    df.drop_duplicates(inplace=True)
    print("Duplicated rows removed.")
else:
    print("No duplicated rows found.")
```

```python
df['flytime_numeric'] = (
    df['flytime'].str.extract(r'(\d+)g (\d+)p')
    .astype(float)
    .apply(lambda x: x[0] * 60 + x[1], axis=1)
)


flytime_range = {
    'min': df['flytime_numeric'].min(),
    'max': df['flytime_numeric'].max()
}

print("Flytime Numeric Values:\n", df[['flytime', 'flytime_numeric']])
print("\nFlytime Range:\n", flytime_range)
```

```python
sorted_dataframe = df.sort_values(by='flytime_numeric', ascending=False)

# Display the sorted DataFrame
print("Sorted DataFrame by Flytime (highest to lowest):\n", sorted_dataframe)
```
